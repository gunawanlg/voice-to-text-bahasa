### Data Preprocessing

#### Split By Symbols

The intuition behind this is to create cleaner transcriptions and divide the transcriptions by symbols that have possibility to create a pause between each word, for this purpose, the following symbols were used `?`, `;`, `:`, `!`. `,` (comma) and `.` (dot) were also used to divide the transcription and since both of them are going to be predicted by the model `[COMMA]` and `[DOT]` sign were used as a way to replace the real symbols so it won't get deleted. Also during the cleaning process these unicode symbols `-`, `—`, `"` were also removed since they're not really used for the voice recognizer. 

```python
def split_by_symbols(string):
    """
    Given a raw string, the function will generate clean string, split it 
    according to the pause symbols and decode `comma` and `dot` as [COMMA], 
    [DOT]
    
    Parameters
    ----------
    string : str
        raw string of transcript obtained from scraping
        
    Returns
    -------
    string : str
        cleaned and splitted transcript based on pause symbols
    """
    string = string.lower()
    
    # Replace the pause symbol with new line symbol, this is done to ease the
    pause_symbols = ["\"\n\n"]
    for pause_symbol in pause_symbols:
        string = string.replace(pause_symbol, "\n")
        
    # Replace all of the - in front of pronouns
    pronouns = {"-ku": "ku", "-nya": "nya"}
    for k, v in pronouns.items():
        string = string.replace(k, v)
    
    # Replace the junk symbols with a whitespace
    junk_symbols = ["-"]
    for junk_symbol in junk_symbols:
        string = string.replace(junk_symbol, " ")
    
    # Replace the junk symbols with empty whitespace
    junk_symbols = [")", "(", "[", "]", "\'", "&gt;", "&lt;", "—"]
    for junk_symbol in junk_symbols:
        string = string.replace(junk_symbol, "")
    
    # Replace umlaut e (ë) with english letter e
    string = string.replace("ë", "e")
    
    # Replace `s e l a` with sela since it's read as a whole
    string = string.replace("s e l a", "sela")
        
    # Replace with possible combinations of comma, dot and newline
    dot_comma_dict = {".": "[DOT]", "!": "[DOT]", "?": "[DOT]", ",": "[COMMA]", ":" : "[COMMA]", ";" : "[COMMA]"}
    for k, v in dot_comma_dict.items():
        for newline in ["\n\n", "\n", " ", ""]:
            string = string.replace(f"{k}{newline}", f"{v}\n")
            
    string = string.replace("\"", "").split("\n")
    string = [s.strip() for s in string]
    
    return string
```

#### Fix String Length

This part is the continuation of the previous preprocessing process. During this part the length of the strings will be fixed so they don't differ that far from each other. The reason behind the creation of this is the fact that if the word/char length is so different/so far from each other, the one with the shortest string will get unnecessary padding. For this purpose two method is proposed: 
1. **Constant Fix String Length** - In constant fix string length the string will be appended by the succeeding string until it reaches a length greater than the `lower_bound`
2. **Smart Fix String Length** - In smart fix string length the string will be appended dynamically comparing the absolute value of the differences between the words and the stated `limit`

##### Constant Fix String Length

```python
def fix_string_length(sentences, lower_bound=10):
    """
    Given a raw string, the function will generate clean string, split it 
    according to the pause symbols and decode `comma` and `dot` as [COMMA], 
    [DOT]
    
    Parameters
    ----------
    sentences : list of str
        list containing clean split strings of transcription
        
    Returns
    -------
    new_sentences : list of str
        better splitted string based on the `lower_bound`
    """
    
    # Create an empty list to store the generated new sentences
    new_sentences = []
    i = 0
    while i < len(sentences) - 1:
        new_sentence = []
        while len(new_sentence) < lower_bound:
            next_sentence = sentences[i].split(" ")
            
            # Ignore the blank character (just in case)
            if next_sentence != [""]:
                new_sentence += next_sentence
            
            # Iterate to the next sentences
            if i < len(sentences) - 1:
                i += 1
            else:
                break
                
        new_sentences.append(" ".join(new_sentence).strip().replace("[DOT]", ".").replace("[COMMA]", ","))
    return new_sentences
```

##### Smart Fix String Length

```python
def smart_fix_string_length(sentences, limit=10):
    """
    Given a raw string, the function will generate clean string, split it 
    according to the pause symbols and decode `comma` and `dot` as [COMMA], 
    [DOT]
    
    Parameters
    ----------
    sentences : list of str
        list containing clean split strings of transcription
        
    Returns
    -------
    new_sentences : list of str
        better splitted string based on the `lower_bound`
    """
    
    new_sentences = []
    i = 0
    
    while i < len(sentences) - 1:
        # Create a bool to control the flow of the loop
        sentence_created = False
        last_sentence_left = False
        new_sentence = []
        
        while not sentence_created:
            next_sentence = sentences[i].split(" ")
            
            # Define the absolute difference between the length of generated 
            # new_sentence and the stated limit
            abs_new = abs(len(new_sentence) - limit)
            
            # Define the absolute difference between the length of generated 
            # new_sentence summed with the next sentence and the stated limit
            abs_new_next = abs(len(new_sentence) + len(next_sentence) - limit)
            
            # Change the bool flag if the sentence fulfills the condition and
            # add the next sentence to the current sentence storage
            if abs_new <= abs_new_next and len(new_sentence) != 0:
                sentence_created = True
                
                if i == len(sentences) - 1:
                    last_sentence_left = True
                break
            else:
                next_sentence = [s for s in next_sentence if s != ""]
                new_sentence += next_sentence
                if i < len(sentences) - 1:
                    i += 1
                else:
                    break
        new_sentence = " ".join(new_sentence).strip().replace("[DOT]", ".").replace("[COMMA]", ",")
        new_sentences.append(new_sentence)
        
        if last_sentence_left:
            new_sentences.append(" ".join(next_sentence).strip().replace("[DOT]", ".").replace("[COMMA]", ","))
    
    # Final cleaning
    new_sentences = [new_sentence for new_sentence in new_sentences if new_sentence != ""]
    
    return new_sentences
```
